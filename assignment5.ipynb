```r
# Function to generate a random string of specified length
generate_random_string <- function(length) {
  chars <- c(letters, LETTERS, 0:9)
  paste0(sample(chars, length, replace = TRUE), collapse = "")
}

# Generate 1000 random strings
random_strings <- replicate(1000, generate_random_string(10))

# Write the random strings to a file
writeLines(random_strings, "random_strings.txt")
```

```r
# Set seed for reproducibility
set.seed(123)

# Create a random dataset of 100 rows and 30 columns
data <- matrix(sample(1:200, 100 * 30, replace = TRUE), nrow = 100, ncol = 30)

# Convert to data frame
df <- as.data.frame(data)

# Replace values between 10 and 60 with NA
df[df >= 10 & df <= 60] <- NA

# Count number of rows having missing values
num_rows_with_na <- sum(apply(df, 1, function(row) any(is.na(row))))
print(paste("Number of rows with missing values:", num_rows_with_na))
```

```r
# Replace NA values with column mean
for (i in seq_along(df)) {
  col_mean <- mean(df[[i]], na.rm = TRUE)
  df[[i]][is.na(df[[i]])] <- col_mean
}
```

```r
# Load required library
library(ggplot2)
library(reshape2)

# Calculate Pearson correlation matrix
cor_matrix <- cor(df)

# Plot heat map of the correlation matrix
cor_heatmap <- melt(cor_matrix)
ggplot(cor_heatmap, aes(Var1, Var2, fill = value)) +
  geom_tile() +
  scale_fill_gradient2(low = "blue", high = "red", mid = "white", midpoint = 0, limit = c(-1, 1)) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1)) +
  labs(title = "Correlation Heatmap", fill = "Correlation")

# Select columns with correlation <= 0.7
selected_columns <- colnames(df)[apply(cor_matrix, 2, function(x) all(x <= 0.7))]
print("Selected columns with correlation <= 0.7:")
print(selected_columns)
```

```r
# Normalize function
normalize <- function(x) {
  return ((x - min(x)) / (max(x) - min(x)) * 10)
}

# Normalize all values in the dataset
df_normalized <- as.data.frame(lapply(df, normalize))
```

```r
# Install required packages if not already installed
if (!requireNamespace("ggplot2", quietly = TRUE)) {
  install.packages("ggplot2")
}
if (!requireNamespace("cluster", quietly = TRUE)) {
  install.packages("cluster")
}
if (!requireNamespace("factoextra", quietly = TRUE)) {
  install.packages("factoextra")
}
if (!requireNamespace("NbClust", quietly = TRUE)) {
  install.packages("NbClust")
}

# Load required libraries
library(ggplot2)
library(cluster)
library(factoextra)
library(NbClust)

# Set seed for reproducibility
set.seed(123)

# Create the random dataset
data <- data.frame(
  col1 = runif(500, -10, 10),
  col2 = runif(500, -10, 10),
  col3 = runif(500, -10, 10),
  col4 = runif(500, -10, 10),
  col5 = runif(500, 10, 20),
  col6 = runif(500, 10, 20),
  col7 = runif(500, 10, 20),
  col8 = runif(500, 10, 20),
  col9 = runif(500, -100, 100),
  col10 = runif(500, -100, 100)
)

# (i) K-Means clustering

# Determine the optimal number of clusters using the Elbow method
fviz_nbclust(data, kmeans, method = "wss") +
  geom_vline(xintercept = 3, linetype = 2) +
  labs(subtitle = "Elbow method for determining optimal number of clusters")

# Perform K-Means clustering with the optimal number of clusters (e.g., 3)
set.seed(123)
kmeans_result <- kmeans(data, centers = 3, nstart = 25)

# Plot the K-Means clustering result
fviz_cluster(kmeans_result, data = data) +
  theme_minimal() +
  labs(title = "K-Means Clustering")

# (ii) Hierarchical clustering

# Calculate the distance matrix
dist_matrix <- dist(data, method = "euclidean")

# Perform hierarchical clustering using Ward's method
hclust_result <- hclust(dist_matrix, method = "ward.D2")

# Determine the optimal number of clusters using the silhouette method
fviz_nbclust(data, hcut, method = "silhouette") +
  labs(subtitle = "Silhouette method for determining optimal number of clusters")

# Cut the dendrogram at the optimal number of clusters (e.g., 3)
cutree_result <- cutree(hclust_result, k = 3)

# Plot the dendrogram
fviz_dend(hclust_result, k = 3, rect = TRUE, rect_fill = TRUE, rect_border = "jco", show_labels = FALSE) +
  theme_minimal() +
  labs(title = "Hierarchical Clustering Dendrogram")

# Plot the hierarchical clustering result
fviz_cluster(list(data = data, cluster = cutree_result)) +
  theme_minimal() +
  labs(title = "Hierarchical Clustering")
```

```r
# Install required packages if not already installed
if (!requireNamespace("ggplot2", quietly = TRUE)) {
  install.packages("ggplot2")
}
if (!requireNamespace("gridExtra", quietly = TRUE)) {
  install.packages("gridExtra")
}

# Load required libraries
library(ggplot2)
library(gridExtra)

# Set seed for reproducibility
set.seed(123)

# Create the random dataset
data <- as.data.frame(matrix(runif(600 * 15, -100, 100), nrow = 600, ncol = 15))

# (i) Plot scatter graph between Column 5 and Column 6
scatter_plot <- ggplot(data, aes(x = V5, y = V6)) +
  geom_point(color = 'blue') +
  labs(title = "Scatter Plot between Column 5 and Column 6", x = "Column 5", y = "Column 6") +
  theme_minimal()

print(scatter_plot)

# (ii) Plot histogram of each column in a single graph
histogram_plots <- lapply(1:15, function(i) {
  ggplot(data, aes_string(x = paste0("V", i))) +
    geom_histogram(binwidth = 10, fill = "blue", color = "black", alpha = 0.7) +
    labs(title = paste("Histogram of Column", i), x = paste("Column", i), y = "Frequency") +
    theme_minimal()
})

# Arrange the histograms in a grid
do.call(grid.arrange, c(histogram_plots, ncol = 3))

# (iii) Plot the Box plot of each column in a single graph
data_long <- reshape2::melt(data)
boxplot <- ggplot(data_long, aes(x = variable, y = value)) +
  geom_boxplot(fill = "lightblue", color = "darkblue") +
  labs(title = "Box Plot of Each Column", x = "Columns", y = "Values") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))

print(boxplot)
```

```r
# Install required packages if not already installed
if (!requireNamespace("ggplot2", quietly = TRUE)) {
  install.packages("ggplot2")
}

# Load required libraries
library(ggplot2)

# Set seed for reproducibility
set.seed(123)

# Create the random dataset
data <- as.data.frame(matrix(runif(500 * 5, 5, 10), nrow = 500, ncol = 5))
colnames(data) <- paste("V", 1:5, sep="")

# (i) Perform t-Test on each column
t_test_results <- lapply(data, function(column) {
  t.test(column, mu = mean(column))  # Perform t-test assuming population mean equals sample mean
})

# Print t-Test results
t_test_results

# (ii) Perform Wilcoxon Signed Rank Test on each column
wilcox_test_results <- lapply(data, function(column) {
  wilcox.test(column, mu = median(column))  # Perform Wilcoxon test assuming population median equals sample median
})

# Print Wilcoxon Signed Rank Test results
wilcox_test_results

# (iii) Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4
# Split the data into two groups for Columns 3 and 4
group1 <- data$V3
group2 <- data$V4

# Perform Two Sample t-Test
two_sample_t_test_result <- t.test(group1, group2)
print("Two Sample t-Test Result:")
print(two_sample_t_test_result)

# Perform Wilcoxon Rank Sum Test (equivalent to Mann-Whitney U test)
wilcoxon_rank_sum_test_result <- wilcox.test(group1, group2)
print("Wilcoxon Rank Sum Test Result:")
print(wilcoxon_rank_sum_test_result)
```
